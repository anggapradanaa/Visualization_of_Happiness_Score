# Menggabungkannya menjadi satu visualisasi

```python
# Mengatur ukuran dan layout untuk multiple subplot
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 12))  # 2 baris dan 3 kolom

# Daftar dataframes dan judul subplot
dataframes = [df_asean, df_east_asia, df_south_asia, df_west_asia, df_central_asia]
titles = ['Wilayah Asia Tenggara', 'Wilayah Asia Timur', 'Wilayah Asia Selatan', 'Wilayah Asia Barat', 'Wilayah Asia Tengah']

# Loop melalui setiap subplot dan dataframe
for i, (df, title) in enumerate(zip(dataframes, titles)):
    row = i // 3
    col = i % 3
    
    # Plot bar untuk setiap wilayah
    ax = sns.barplot(x='Country', y='Happiness score', data=df, palette='crest', ax=axes[row, col])
    ax.set_title(f'Happiness Score berdasarkan {title}')
    ax.set_xlabel('Negara')
    ax.set_ylabel('Happiness Score')
    ax.tick_params(axis='x', rotation=15)  # Mengatur rotasi label sumbu x
    
    # Tambahkan garis putus-putus untuk menandakan skor kebahagiaan rata-rata (Global Mean)
    ax.axhline(mean_score, linestyle='--', color='red', label='Global Mean')

    # Menambahkan keterangan nilai Happiness Score di atas setiap bar
    for index, value in enumerate(df['Happiness score']):
        ax.text(index, value + 0.1, f'{value:.2f}', ha='center', va='bottom', fontsize=9)

    # Hapus grid dan spines (garis tepi) yang tidak perlu
    ax.grid(False)
    sns.despine(ax=ax, left=True, bottom=True)
    
    # Tampilkan legend di setiap subplot
    ax.legend(loc='upper right', fontsize='small')

# Hapus subplot yang tidak digunakan
fig.delaxes(axes[1, 2])

plt.tight_layout()  # Mengatur layout subplot agar rapi
plt.show()
```

# Perbandingan antara negara dengan Happiness Score tertinggi dari masing-masing wilayah

```python
# Mengambil negara dengan happiness score tertinggi dari masing-masing wilayah
top_country_asean = df_asean.loc[df_asean['Happiness score'].idxmax()]
top_country_east_asia = df_east_asia.loc[df_east_asia['Happiness score'].idxmax()]
top_country_south_asia = df_south_asia.loc[df_south_asia['Happiness score'].idxmax()]
top_country_west_asia = df_west_asia.loc[df_west_asia['Happiness score'].idxmax()]
top_country_central_asia = df_central_asia.loc[df_central_asia['Happiness score'].idxmax()]

# Membuat dataframe baru untuk menampung data negara dengan happiness score tertinggi dari setiap wilayah
top_countries_asia_df = pd.DataFrame({
    'Region': ['Asia Tenggara', 'Asia Timur', 'Asia Selatan', 'Asia Barat', 'Asia Tengah'],
    'Country': [top_country_asean['Country'], top_country_east_asia['Country'],
                top_country_south_asia['Country'], top_country_west_asia['Country'],
                top_country_central_asia['Country']],
    'Happiness Score': [top_country_asean['Happiness score'], top_country_east_asia['Happiness score'],
                        top_country_south_asia['Happiness score'], top_country_west_asia['Happiness score'],
                        top_country_central_asia['Happiness score']],
    'Rank': [top_country_asean['RANK'], top_country_east_asia['RANK'],
             top_country_south_asia['RANK'], top_country_west_asia['RANK'],
             top_country_central_asia['RANK']]
})

# Mengurutkan DataFrame berdasarkan Happiness Score secara menurun
top_countries_asia_df = top_countries_asia_df.sort_values(by='Happiness Score', ascending=False)

# Mengatur ukuran figur grafik
plt.figure(figsize=(12, 8))

# Membuat plot untuk menampilkan negara dengan happiness score tertinggi dari masing-masing wilayah
ax = sns.barplot(x='Region', y='Happiness Score', hue='Country', data=top_countries_asia_df, palette=colors_cream, dodge=False)

# Menambahkan nilai Happiness Score dan Rank pada setiap bar
for i, v in enumerate(top_countries_asia_df['Happiness Score']):
    plt.text(i, v + 0.25, f'{v:.2f}', ha='center', color='black', fontweight='bold')
    plt.text(i, v - 0.5, f'Rank: {top_countries_asia_df.iloc[i]["Rank"]}', ha='center', color='black', fontstyle='italic')

# Menetapkan batasan rentang sumbu y agar lebih luas
plt.ylim(0, top_countries_asia_df['Happiness Score'].max() + 0.75)    

# Mendapatkan batas-batas sumbu x dan y
xmin, xmax = ax.get_xlim()
ymin, ymax = ax.get_ylim()

# Tambahkan garis putus-putus untuk menandakan skor kebahagiaan rata-rata (Global Mean)
plt.axhline(mean_score, linestyle='--', color='red', label='Global Mean')

# Tambahkan teks "Global Mean" di luar batas sumbu x
ax.text(xmax + 0.1, mean_score, f'Global Mean: {mean_score:.2f}', ha='left', va='center', fontsize=12, color='red')

plt.title('Negara dengan Happiness Score Tertinggi dari Setiap Wilayah Asia', fontsize=14)
plt.xlabel('Wilayah', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)
plt.legend(title='Negara', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.tight_layout()  # Mengatur layout agar rapi
plt.show()
```

```python
# Filter rows for Iraq and Indonesia
countries_third = ['Indonesia', 'Iraq']


# Columns to include in the radar chart
columns = ['Dystopia (1.83) + residual', 'Explained by: GDP per capita',
           'Explained by: Social support', 'Explained by: Healthy life expectancy',
           'Explained by: Freedom to make life choices',
           'Explained by: Generosity', 'Explained by: Perceptions of corruption']

# Number of variables we're plotting.
num_vars = len(columns)

# Compute angle each bar is centered on:
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # Complete the loop by appending the first angle at the end.

# Draw the radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

for k, (idx, row) in enumerate(df_plot_third.iterrows()):
    data = row[columns].tolist() + row[columns].tolist()[:1]  # Complete the loop
    ax.fill(angles, data, alpha=0.4, label=row['Country'], color=colors_green[3*k], edgecolor=colors_green[3*k])
    ax.plot(angles, data)

# Labels for each point
ax.set_xticks(angles[:-1])
col_name = ['Dystopia', 'GDP per capita',
           'Social support', 'Healthy life expectancy',
           'Freedom to make life choices',
           'Generosity', 'Perceptions of corruption']
ax.set_xticklabels(col_name)

# Title and legend
plt.title('Comparison of Happiness Factors: Iraq vs. Indonesia', size=15, color=colors_green[1], y=1.1)
ax.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))
plt.show()
```

```python
# Filter rows for Indonesia and Iraq
countries_third = ['Indonesia', 'Iraq']
df_plot_third = df_third_countries[df_third_countries['Country'].isin(countries_third)]

# Columns to include in the stacked bar chart
columns = ['Dystopia (1.83) + residual', 'Explained by: GDP per capita',
           'Explained by: Social support', 'Explained by: Healthy life expectancy',
           'Explained by: Freedom to make life choices',
           'Explained by: Generosity', 'Explained by: Perceptions of corruption']

# Set data for each country
category_indo = df_plot_third[df_plot_third['Country'] == 'Indonesia'][columns].values.flatten()
category_iraq = df_plot_third[df_plot_third['Country'] == 'Iraq'][columns].values.flatten()

# Set the labels for each category
labels = columns

# Set the positions for the bars
x = np.arange(len(labels))

# Set the width of the bars
width = 0.35

# Plot the stacked bars
fig, ax = plt.subplots(figsize=(12, 8))

# Plot bars for Indonesia
bars_indo = ax.barh(x - width/2, category_indo, width, label='Indonesia', color='skyblue')

# Plot bars for Iraq on top of Indonesia's bars
bars_iraq = ax.barh(x + width/2, category_iraq, width, label='Iraq', color='lightcoral')

# Add labels to the bars (Indonesia)
for i, (category, value) in enumerate(zip(labels, category_indo)):
    ax.text(value, i - width/2, f'{value:.2f}', ha='left', va='center', color='black')

# Add labels to the bars (Iraq)
for i, (category, value) in enumerate(zip(labels, category_iraq)):
    ax.text(value, i + width/2, f'{value:.2f}', ha='left', va='center', color='black')

# Add labels, title, and legend
ax.set_xlabel('Explained by Score')
ax.set_ylabel('Factors')
ax.set_title('Comparison of Happiness Factors: Iraq vs. Indonesia')
ax.set_yticks(x)
ax.set_yticklabels(labels)
ax.legend()

# Adjust layout
plt.tight_layout()
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)

# Show plot
plt.show()
```
